Seaborn is a statical plotting library that is specifically designed to intract well with Pandas DataFrames to create common statical plot types.

Seaborn is built directly off of Matplotlib bu uses a simpler "one-line" Syntax.

When using seaborn, we trade-off customization for ease of use.

A typical seaborn plot uses one line of code for example:

sns.scatterplot(x='salary', y='sales', data = df)

Seaborn takes in a pandas Dataframe and then user provides the corresponding string column names for x and y (depending on the plot type)

# Scatters Plots with Seaborn

Scatter plots show the relationship between two continuous features.

Scatter plots line up a set of two continous features and plots them out as coordinates. For Example, imageine employees with salaries who sell a certain dollar amount of items each year. We could explore the relationship between employee salaries and sales amount.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df
```

```python
plt.figure(figsize =(12,4),dpi = 200)

sns.scatterplot(x='salary', y = 'sales', data=df)
```

```python
# if we want to seprate the dots on the bases of any catagory than we use hue paramerter of scatterplot
plt.figure(figsize =(12,4), dpi =200)
sns.scatterplot(x='salary', y='sales', data = df, hue = 'level of education')
```

```python
#if we seprate dots based on continous column
plt.figure(figsize =(12,4), dpi =200)
sns.scatterplot(x='salary', y='sales', data = df, hue = 'salary')
```

```python
# figure based on the size of dots
plt.figure(figsize=(12,4), dpi = 200)
sns.scatterplot(x ='salary', y ='sales', data = df, size='salary', alpha=0.5);# alpha means transparency 
```

```python
plt.figure(figsize =(12,4), dpi=200)
sns.scatterplot(x='salary', y ='sales', data =df, s=200, style='level of education', hue='level of education')
```

# Distribution Plots

Understanding Plots type

Distribution plots display a single continuous feature and help visualize properties such as deviation and average values.

There are 3 main distribution plot type.
1) Rug plot
2) Histogram
3) KDE plot

A rug plot is the simplest distribution plot and merely adds a dash or tick line for every single value.

The y-axis does not really have a meaning.

KDE stands for Kernel Density Estimation. 


It is a method of estimating a probability density function of a random variable.

In simpler terms, it is a way of estimating a continous probability curve for a finite data sample

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
# plt.figure(figsize=(4,2), dpi =200)
sns.rugplot(x='salary', data=df, height =0.5)
```

```python
sns.displot(data=df, x='salary', bins =10)
```

```python
sns.displot(data = df, x = 'salary', bins=100)
```

```python
# Styling of Graph
# sns.set(style = 'whitegrid')
sns.displot(data = df, x='salary', bins=20)
```

```python
sns.displot(data = df, x='salary', bins=30, color='black')
```

```python
sns.displot(data = df, x='salary', bins=20, color='red',edgecolor='blue', linewidth = 4)
```

```python
sns.displot(data=df, x='salary', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

# Statistical Estimation within Categories 

Unsderstanding The plots

The catagorical plots discussed here will display a statical metrics per category.

For example mean value per category or a count of the number of rows per category

It is the visualixation equivalent of a groupby() call.

The two main types of plots for this are : - countplot() => Count numbers of rows per category

barplot() => General form of displaying any chosen metric per category.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(data=df, x='division')
# plt.ylim(0, 280)
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(data=df, x='level of education', hue = 'division', palette = 'Set2')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education', y='salary', estimator =np.mean, ci='sd', hue = 'division')

plt.legend(bbox_to_anchor =(1.05,1))
```

# jointplot()
We can map histograms to each feature of a scatterplot to clarify the distribution within each feature.

# pairplot()
The pairplot() is a quick way to compare all numerical columns in a DataFrame.

It automatically creates a histogram for each column and a scatterplot comparison between all possible combination of columns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.jointplot(data=df, x='math score',y = 'reading score', kind ='scatter', alpha =.5)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind ='hist')
```

```python
sns.jointplot(data = df, x='math score', y= 'reading score', kind ='kde', shade =True)
```

# Seaborn Grid

Seaborn grid calls use Matplotlib subplots() to automatically create a grid based off a categorical column.

Instead of passing in a specific number of cols or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid.

Many of seaborn's built-in calls are running on top of this grid system.
Directly calling the grid system allows users to heavily customize plots.

```python
sns.catplot(data=df,x='gender', y='math score', kind = 'box' )
```

```python
sns.catplot(data = df, x='gender', y ='math score', kind ='box',
           row ='lunch')
```

```python
sns.catplot(data = df, x='gender', y = 'math score',kind = 'box', hue='lunch')
```

```python
sns.catplot(data = df, x= 'gender', y = 'math score', kind='box', col = 'lunch')
```

```python
sns.catplot(data = df, x='gender', y = 'math score', kind ='box', 
           col = 'lunch', row ='test preparation course')
```

```python
sns.catplot(data = df, x='gender', y = 'math score', kind= 'violin', 
           col = 'lunch', row='test preparation course')
```

```python
g=sns.PairGrid(df)

g= g.map_upper(sns.scatterplot)

g=g.map_diag(sns.histplot)

g=g.map_lower(sns.kdeplot)
```

# Matrix Plot
These plots are the visual equivalent of displaying a pivot table.
The matrix plot displays all the data passed in visualizing all the numeric values in a DataFrame

The two main matrix plot types are:
1. heatmap() :- Visually displays the distribution of cell values with a color mapping.
2. clustermap() :- Smae visual as heatmap, but first conducts hierarchical clustering to reorganize data into groups

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df.head()
```

```python
df = df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df.drop('Life expectancy',axis = 1))
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth=0.5, annot=True,
           cmap = 'viridis')
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
           cmap='viridis', center = 10)
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy', axis =1), linewidth =0.5, annot =True,
           cmap = 'viridis', col_cluster =False )
```
